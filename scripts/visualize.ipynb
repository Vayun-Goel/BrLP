```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

# Load the .nii.gz file
file_path = '/home/vayub/BrLP/output_inference_imgtoimg/followup_age_70.nii.gz'  # Replace with your actual file path
img = nib.load(file_path)
data = img.get_fdata()

data = np.max(data) - data

# Determine center slices
z, y, x = data.shape
sagittal = data[z // 2, :, :]
coronal = data[:, y // 2, :]
axial = data[:, :, x // 2]


# Plot the slices
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(np.rot90(sagittal), cmap='gray')
axes[0].set_title('Sagittal Slice')

axes[1].imshow(np.rot90(coronal), cmap='gray')
axes[1].set_title('Coronal Slice')

axes[2].imshow(np.rot90(axial), cmap='gray')
axes[2].set_title('Axial Slice')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
